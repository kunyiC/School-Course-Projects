```python
class_label = file1_txt[0]
xcen = file1_txt[1]
ycen = file1_txt[2]
w = file1_txt[3]
h = file1_txt[4]
```

```python
from PIL import Image 
I = Image.open(base_dir+'17103657_20200806_LL.jpg')
print("1. size:", I.size)
```

```python
xmin = max(float(xcen) - float(w) / 2, 0)
xmax = min(float(xcen) + float(w) / 2, 1)
ymin = max(float(ycen) - float(h) / 2, 0)
ymax = min(float(ycen) + float(h) / 2, 1)

xmin = int(I.size[1] * xmin)
xmax = int(I.size[1] * xmax)
ymin = int(I.size[0] * ymin)
ymax = int(I.size[0] * ymax)
```

```python
print(xmin)
print(xmax)
print(ymin)
print(ymax)
```
