#### 參考 DL4b 程式，用 PCA & t-SNE 分析 mnist 資料集的特徵

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt 
import os
%matplotlib inline
```

```python
mnist = load_digits()

X_tsne = TSNE(n_components=2, 
              random_state=33).fit_transform(mnist.data)
```

```python
X_pca = PCA(n_components=2).fit_transform(mnist.data)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=mnist.target)
plt.title("t-SNE of MNIST")

plt.subplot(122)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=mnist.target)
plt.title("PCA of MNIST")

plt.show()
```
