# DL4b Demos of PCA&t-SNE with the iris dataset

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris,load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import os

%matplotlib inline
```

```python
iris = load_iris()
X_tsne = TSNE(n_components=3,random_state=33).fit_transform(iris.data)
X_pca = PCA(n_components=3).fit_transform(iris.data)

ckpt_dir="images"
if not os.path.exists(ckpt_dir):
    os.makedirs(ckpt_dir)

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=iris.target,label="t-SNE")
plt.legend()
plt.subplot(122)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=iris.target,label="PCA")
plt.legend()
plt.show()
```
