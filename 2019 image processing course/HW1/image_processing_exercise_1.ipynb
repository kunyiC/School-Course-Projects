<h2 align="left"><font color=blue>(1) Define functions</font></h2>

```python
from PIL import Image                          
import matplotlib.pyplot as plt   
import matplotlib.image as mpimg
from matplotlib.colors import NoNorm
```

```python
# show image & get its information
# parameters: image_dir, image_title
# parameters: gray_or_not (True)

def show_image(image_dir, image_title, gray_or_not):
    img = mpimg.imread(image_dir)
    if gray_or_not == True:
        plt.imshow(img, cmap='gray',norm=NoNorm())
    else:
        imgplot = plt.imshow(img)
    plt.title(image_title)
    plt.axis('off') 
    plt.show()
    I = Image.open(image_dir)
    print("Image's information:")
    print("1. size:", I.size)
    print("2. mode:", I.mode)
    print("3. format:", I.format)
```

```python
# convert colored image into gray scale image & save it
import cv2

def convert2gray(original_image, converted_image):
    image = cv2.imread(original_image)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    cv2.imwrite(converted_image, gray_image)
```

```python
# calculate psnr (peak signal-to-noise ratio)

import numpy as np
import math

def psnr(original_image, gray_image):
    original = cv2.imread(original_image)
    contrast = cv2.imread(gray_image)
    mse = np.mean((original - contrast)**2)
    if mse == 0:
        return 100
    PIXEL_MAX = 255.0
    print("The psnr is:")
    return (20 * math.log10(PIXEL_MAX / math.sqrt(mse)))
```

<h2 align="left"><font color=blue>(2) Image 1 (Lena):</font></h2>

```python
show_image('lena.jpg', 'Lena: colored', False)
```

```python
convert2gray('lena.jpg', 'lena_gray.jpg')
```

```python
show_image('lena_gray.jpg', 'Lena: gray', True)
```

```python
psnr("lena.jpg", "lena_gray.jpg")
```

```python
psnr("self.bmp", "self.gif")
```
