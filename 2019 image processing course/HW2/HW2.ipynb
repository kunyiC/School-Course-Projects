<h2 align="left"><font color=blue>1. Define functions</font></h2>

```python
from PIL import Image                         
import matplotlib.pyplot as plt   
import matplotlib.image as mpimg
```

```python
# show image & get its information
# parameters: image path, image title, image size, image format

def show_image(img_dir, img_title):
    img = mpimg.imread(img_dir)
    imgplot = plt.imshow(img)
    plt.title(img_title)
    plt.axis('off') 
    plt.show()
    I = Image.open(img_dir)
    print("Image's information:\n")
    print("mode:", I.mode)
    print("format:", I.format)
    print("size:", I.size)
```

<h3 align="left"><font color=purple>(1) Negative transformation:</font></h3>

```python
# invert image
from PIL import ImageOps 

def invert_img(img_dir, img_title):
    img = Image.open(img_dir)
    img_neg = ImageOps.invert(img)
    #img_neg.save(invert_img_dir)
    #inverted_img = mpimg.imread(invert_img_dir)
    plt.imshow(img_neg)
    plt.title(img_title)
    plt.axis('off') 
    plt.show()
```

<h3 align="left"><font color=purple>(2) Sobel edge detection:</font></h3>

```python
import numpy as np
import cv2
from skimage.filters import sobel

def sobel_operator(img_dir, img_title):
    img = cv2.imread(img_dir)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edge_sobel = sobel(img_gray)
    plt.imshow(edge_sobel)
    plt.title(img_title)
    plt.axis('off') 
    plt.show()
```
